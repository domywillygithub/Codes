```python
from random import randint, random
import matplotlib.pyplot as plt
import numpy as np
from numba import jit


probabilities = np.linspace(0.01, 1, 100)
mc_steps = 1000
size = 100


@jit
def simulation(lattice):
    while sum(lattice) != size and sum(lattice) != 0:
        position = randint(0, size - 1)

        if random() >= 0.5:
            lattice[position] = lattice[(position + 1) % size]
        else:
            lattice[position] = lattice[(position - 1) % size]

    return sum(lattice) == size
```

```python
def main():
    results = []
    for pr in probabilities:
        print(pr)
        num_of_successes = 0
        for _ in range(mc_steps):
            lattice = np.random.choice([0, 1], size=size, p=[1.0 - pr, pr])
            num_of_successes += simulation(lattice)
        results.append(num_of_successes / mc_steps)
    print(results)

    plt.plot(probabilities, results)
    plt.show()


if __name__ == '__main__':
    main()
```
