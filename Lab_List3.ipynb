```python
import numpy as np

#define Jaccard Similarity function
def Jaccard():
    a = {0, 1, 2, 5, 6, 8, 9}
    b = {0, 2, 3, 4, 5, 7, 9}
    for i in a or b:
        if len(a) == 0 or len(b) == 0:
            print("There exist an empty set(s)")
            break
    else:
        print("None of the sets is empty")
    intersection = len(set(a).intersection(b))
    union = (len(a) + len(b)) - intersection
    return (intersection) / union

#find Jaccard Similarity between the two sets 
Jaccard()
```

```python
import string
import nltk
from nltk.corpus import gutenberg

tokens = dict()
for i in range(1,10):
    text = eval(f"text{i}")
    text = str([w for w in text]).translate(str.maketrans('', '', string.punctuation))
    tokens[f"{i}"] = nltk.word_tokenize(text)
```

```python
similarity_dict = dict()

for k, v in tokens.items():
    for k_, v_ in tokens.items():
        if k != k_:
            similarity_dict[f"{k}-{k_}"] = nltk.jaccard_distance(set(v), set(v_))    
```

```python
similar_consecutive_words = dict()
for i, word in enumerate(tokens["1"]):
    if i < len(tokens["1"]) - 1:
        similar_consecutive_words[f"{word}-{tokens['1'][i+1]}"] = nltk.jaccard_distance(set(word), set(tokens["1"][i+1]))
```

```python
def comp_gray(b, b_prime, gc_size, n=2):
    result = dict()
    dist = 0
    for i, b_p in enumerate(b_prime):
        dist = distance.hamming(split(b), split(b_p)) * gc_size
        if dist == n:
            print(f"{b} - {b_p} distance: ", dist)
            result[i] = b_p

    return result
```

```python
def get_shingles(text, k = 3):
    for i in range(len(text) - k+1):
        yield text[i:i+k]
```

```python
list(get_shingles("abcdefghijklmnopqrstuvwxyz"))
```

```python
def create_shingles():
    doc = input("Enter a string: ")
    k = int(input("Enter number of shingles: "))
    
    doc.split()
    doc_length = len(doc) 
    #print(doc_length)
    shingled_set = set()
    # iterate through the string and slice it up by k-chars at a time
    for idx in range(doc_length - k + 1):
        doc_slice = doc[idx:idx + k]
        shingled_set.add(doc_slice)

    print(shingled_set)

create_shingles()
```
