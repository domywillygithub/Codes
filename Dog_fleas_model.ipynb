```python
import pylab
import random

#Given two dogs
ACE = []
REX = []
N = [0,10**6]
tsym = [0,10**6]
P = [0,1]

#Initialize the number of fleas i.e N=100 fleas

N = 100
#for whichflea present initially append it to dog 1(ACE)

for i in range(N):
    ACE.append(True)
    REX.append(False)
    
#initialize the number of fleas count between the two dogs(NA for ACE and NR for REX)       
NA = [] 
NR = []
p = 0.6
#The number of times it takes to run the model 

times = 1000

# Open a text file for writing
with open(f"N{N}p{p}t{times}.txt", "w") as file:
    # Write the values of N, p, and times to the file
    file.write(f"N={N}\n")
    file.write(f"p={p}\n")
    file.write(f"times={times}\n")

    for time in range(times):
        i = random.randint(0,N-1)  #select a random number of flea to jump in between two dogs with a given probability (p=0.3)
        
        P  = random.uniform(0, 1)
        if P <= 0.6:   # you can change the probability to observe the model 
            
            if ACE[i]:                        #if fleas are already in ACE then they should jump to REX
                ACE[i]=False

                REX[i]=True
            else:
                REX[i]=False         #if fleas are already in REX then they should jump to ACE 

                ACE[i]=True 
            
        #check where are the fleas present     
        flea_ACE = 0
        flea_REX = 0
        for fleaexist in ACE:
            if fleaexist:
                flea_ACE = flea_ACE + 1
            else:
                flea_REX = flea_REX + 1
        NA.append(flea_ACE)
        NR.append(flea_REX)
        
    # Write the values of NA and NR to the file
    file.write("NA=" + str(NA) + "\n")
    file.write("NR=" + str(NR) + "\n")
    
pylab.plot(NA)
```
