```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/book1.xlsx')
df.head()
```

```python
districts = df['District'].unique()

for district in districts:
    district_data = df[df['District'] == district]

    plt.figure(figsize=(10, 6))
    plt.plot(district_data['Year'], district_data['Suspected'], label='Suspected Cases', marker='o')
    plt.plot(district_data['Year'], district_data['Confirmed cases'], label='Confirmed Cases', marker='o')

    plt.title(f'Malaria Cases Over Years in {district}')
    plt.xlabel('Year')
    plt.ylabel('Number of Cases')
    plt.legend()
    plt.grid(True)
    plt.show()
```
