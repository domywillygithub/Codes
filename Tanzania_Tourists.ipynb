# Tourists Visiting Tanzania from 1995 to 2019

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
df  = pd.read_csv('/content/Tourists.csv', index_col = 'Year')
```

```python
df['Number of Tourists in Million']=df['Number of Tourists'].values/1000000
df
```

```python
sns.set()
plt.figure(figsize=(10,5))

ax = sns.lineplot(data=df, x='Year', y='Number of Tourists in Million', marker = 'o')
plt.title('Number of Tourists visiting Tanzania from 1995 to 2019')
plt.xlabel('Year')
plt.ylabel('Number of Tourists (in millions)')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# create a dataframe with dates from 2010 to 2020 and random values for the number of tourists
dates = pd.date_range(start='2010-01-01', end='2020-12-31', freq='M')
num_tourists = np.random.randint(500000, 2000000, size=len(dates))
df = pd.DataFrame({'date': dates, 'num_tourists': num_tourists})

# extract month and year information from the date column
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

# create a pivot table to group the data by year and month
pivot_df = df.pivot_table(index='month', columns='year', values='num_tourists')

# plot the seasonal time series
plt.figure(figsize=(10, 5))
sns.set_style("darkgrid")
sns.lineplot(data=pivot_df, markers='o')
plt.title('Number of Tourists visiting Tanzania by Season')
plt.xlabel('Month')
plt.ylabel('Number of Tourists')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# create a larger dataset
dates = pd.date_range(start='2015-01-01', end='2022-12-31', freq='M')
data = np.random.randint(1000, 5000, size=len(dates))
df = pd.DataFrame({'date': dates, 'value': data})

# plot the data
plt.figure(figsize=(15,6))
sns.set()
ax = sns.lineplot(data=df, x='date', y='value', marker='o')
plt.title('Total number of weddings across different regions in the US from 2010 to 2022')
plt.ylabel('Number of Weddings')
plt.xlabel('Year')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create example data
np.random.seed(123)
dates = pd.date_range(start='2010-01-01', end='2021-12-31', freq='D')
values = np.sin(np.linspace(0, 2*np.pi, len(dates))) + np.random.normal(0, 0.2, len(dates))
df = pd.DataFrame({'Date': dates, 'Value': values})

# Add monthly seasonality
df['Month'] = df['Date'].dt.month
df['Value'] += 0.5*np.sin(2*np.pi*df['Month']/12) * 100

# Plot data
plt.figure(figsize=(15,6))
sns.set()
ax = sns.lineplot(data=df, x='Date', y='Value')
plt.title('The number of wedding across different regions in the US from 2010 to 2022')
plt.xlabel('Year')
plt.ylabel('Number of Wedding')
plt.show()
```
