```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
sns.set()
```

```python
df  = pd.read_csv('/content/force2020_data_unsupervised_learning.csv', index_col='DEPTH_MD')
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.describe()
```

```python
scaler = StandardScaler()
```

```python
df[['RHOB_T', 'NPHI_T', 'GR_T','PEF_T','DTC_T']] = scaler.fit_transform(df[['RHOB','NPHI','GR','PEF','DTC']])
```

```python
df
```

```python
def optimise_k_means(data, max_k):
  means = []
  inertia = []

  for k in range(1,max_k):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data) 

    means.append(k)
    inertia.append(kmeans.inertia_)

#Generate the elbow plot
  fig = plt.subplots(figsize=(12,10))
  plt.plot(means,inertia, 'o-')
  plt.xlabel('Number of clusters')
  plt.ylabel('Inertia')
  plt.grid(True)
  plt.show()
```

```python
optimise_k_means(df[['RHOB_T', 'NPHI_T']],10)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(df[['NPHI_T','RHOB_T']])
```

```python
df['kmeans_3'] = kmeans.labels_
```

```python
df
```

```python
plt.scatter(x = df['NPHI'], y=df['RHOB'], c = df['kmeans_3'], cmap='rainbow')
plt.xlim(-0.1,1)
plt.ylim(3,1.5)
plt.show()
```
