```python
import requests
from bs4 import BeautifulSoup
import random


def scrapeWikiArticle(url, n):
    i = 0
    while i < n:
        response = requests.get(
            url=url,
        )

        soup = BeautifulSoup(response.content, 'html.parser')

        allLinks = soup.find(id="bodyContent").find_all("a")
        random.shuffle(allLinks)
        linkToScrape = 0

        valid_links = []
        for link in allLinks:
            # We are only interested in other wiki articles
            try:
                if link['href'].find("/wiki/") == -1:
                    continue
                if link['href'].find(":") > -1:
                    continue
                if link['href'].find(".") > -1:
                    continue
            except:
                continue
            # Use this link to scrape
            print(i, link['href'].replace("/wiki/", ""))
            valid_links.append(link)
            linkToScrape = link



        curr_link = url.replace("https://en.wikipedia.org/wiki/", "")
        with open('pages.txt', 'a') as f:
            for link in valid_links:
                connected_link = link['href'].replace("/wiki/", "")
                try:
                    f.write(f'{curr_link}\t{connected_link}\n')
                except:
                    pass


        url = "https://en.wikipedia.org" + linkToScrape['href']
        i += 1


scrapeWikiArticle("https://en.wikipedia.org/wiki/machine_learning", 10)
```
